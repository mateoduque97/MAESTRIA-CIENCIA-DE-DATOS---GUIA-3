```python
import pandas as pd

# Leer el archivos
df1 = pd.read_csv('datos_secop_I_modified.csv', sep=',')
df2 = pd.read_csv('datos_secov_II_modified.csv', sep=';', encoding='latin-1')
```

# Columnas a concatenar de ambos dataframes.

Identificamos y agrupamos en una variable las columnas que se relacionan entre ambas bases de datos distintivamente para posteriormente poder concatenarlas y asi poder realizar el analisis respectivo.

```python
# Seleccion de columnas
columnas_df1 = ['Nombre Entidad', 'NIT', 'Año', 'Presupuesto Total', 'Limite Menor Cuantia', 'Limite Minima Cuantia', 'Fecha ultima actualizacion'
                , 'Fecha de Cargue', 'Municipio', 'Departamento']
columnas_df2 = ['Nombre Entidad', 'NIT', 'Año', 'Presupuesto Total', 'Limite Menor Cuantia', 'Limite Minima Cuantia', 'Fecha ultima actualizacion'
                , 'Fecha de Cargue', 'Municipio', 'Departamento']

df1_seleccionadas= df1[columnas_df1]
df2_seleccionadas = df2[columnas_df2]

# Concatenar DataFrames verticalmente (por defecto axis=0)
df3 = pd.concat([df1_seleccionadas, df2_seleccionadas],ignore_index=True)

# Mostrar las primeras filas del DataFrame combinado
df3.head()
```

## **Limpieza de datos**

```python
#Limpieza de Nit
#Quitar espacios y guiones
df3['NIT'] = df3['NIT'].astype(str).str.replace(' ', '').str.replace('-', '')

# Ajustar los NITs: si tienen más de 9 caracteres, quitar los últimos caracteres
df3['NIT'] = df3['NIT'].apply(lambda x: x[:9] if len(x) > 9 else x)

# Contar el número de caracteres en cada NIT
df3['NIT_length'] = df3['NIT'].str.len()

print(df3['NIT_length'])
```

```python
df3['NIT'].str.len().unique() # Como podemos observar todos los NIT cumplen la caracteristica de tener 9 caracteres.
```

```python
#Quitar signos de $ en columanas de montos
columnas_a_limpiar = ['Presupuesto Total', 'Limite Menor Cuantia', 'Limite Minima Cuantia']  # Reemplaza con los nombres de tus columnas

# Quitar los signos de pesos de las columnas especificadas
for columna in columnas_a_limpiar:
    df3[columna] = df3[columna].astype(str).str.replace('$', '').str.replace(',', '')
    
df3
```

```python
#Eliminar filas nulas
# Contar el número de filas antes de limpiar
num_filas_antes = len(df3)

#Quitar valores nulos y de texto de las columnas a limpiar
df3 = df3.dropna(subset=columnas_a_limpiar)

# Contar el número de filas después de limpiar
num_filas_despues = len(df3)

# Calcular el número de filas eliminadas
filas_eliminadas = num_filas_antes - num_filas_despues

# Mostrar el número de filas eliminadas
print(f'Número de filas eliminadas: {filas_eliminadas}')
```

# Determinar cuales son los datos que seran el medio a analizar

Determinamos que el analisis de datos se realizara en las variables numericas y la limpieza se realiza en base a estas para todo el dataframe concatenado.

```python
# Contar el número de filas
num_filas_antes = len(df3)

# Para almacenar las filas que no se pudieron convertir
filas_no_convertidas = pd.DataFrame()

# Intentar convertir las columnas a tipo numérico y detectar filas no convertibles
for columna in columnas_a_limpiar:
    df3[columna] = pd.to_numeric(df3[columna], errors='coerce')  # 'coerce' convierte valores inválidos a NaN
    # Almacenar las filas con NaN en esa columna
    filas_no_convertidas = pd.concat([filas_no_convertidas, df3[df3[columna].isna()]])

# Eliminar las filas con NaN en las columnas especificadas
df3 = df3.dropna(subset=columnas_a_limpiar)

# Contar el número de filas después de la limpieza
num_filas_despues = len(df3)

# Calcular el número de filas eliminadas
filas_eliminadas = num_filas_antes - num_filas_despues

# Mostrar el número de filas eliminadas
print(f'Número de filas eliminadas: {filas_eliminadas}')
```

## **Analisis y Graficos**

```python
#presupuesto total por Año
presupuesto_total = df3.groupby(['Año'])['Presupuesto Total'].sum()
print(presupuesto_total)
```

```python
#Grafica por año
import matplotlib.pyplot as plt

data_agrupada = df3.groupby('Año')['Presupuesto Total'].sum()

# Crear la gráfica de barras
plt.figure(figsize=(7, 5))
data_agrupada.plot(kind='bar', color='skyblue')

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Presupuesto Total')
plt.title('Presupuesto Total por Año')

# Mostrar la gráfica
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### Se logra evidenciar que en el 2021, y 2022 fueron los años donde menos presupuesto total hubo

```python
# Reemplazar 'Bogotá' por 'Bogotá D.C.' en la columna 'Ciudad'
df3['Departamento'] = df3['Departamento'].replace('Bogotá D.C.', 'Bogotá', regex=True)

#Presupuesto Total por Departamento
presupuesto_dep = df3.groupby(['Departamento'])['Presupuesto Total'].sum()
print(presupuesto_dep)
```

```python
# Crear la gráfica de líneas
plt.figure(figsize=(25, 6))
plt.plot(presupuesto_dep.index, presupuesto_dep.values, marker='o', color='skyblue', linestyle='-', linewidth=2)

# Añadir etiquetas y título
plt.xlabel('Departamento')
plt.ylabel('Suma de Valores')
plt.title('Suma de Valores por Departamento')

# Mostrar la gráfica
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

##### Se logra apreciar que la mayoria del presupuesto total pertenece a las empresas de antioquia

```python
# Graficar una torta para el 2024
data_2024 = df3[df3['Año'] == 2024].iloc[0]
labels = ['Limite Minima Cuantia', 'Limite Menor Cuantia']
sizes = [data_2024['Limite Minima Cuantia'], data_2024['Limite Menor Cuantia']]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Configurar la gráfica
plt.title('Distribución del Presupuesto en 2024')

# Mostrar la gráfica
plt.show()
```
